#LIGHT CURVES FOR SEE-CHANGE

#SNe Type Ia 

```python
%matplotlib inline
import lightcurves
from IPython.html.widgets import interact, FloatSliderWidget, fixed

interact(lightcurves.plot_Ia, z = FloatSliderWidget(min=0.15, max=2, step=0.05, value=1.60),
         x1 = FloatSliderWidget(min=-1, max=1, step=0.1, value=0),
         c = FloatSliderWidget(min=-0.1, max=0.1, step=0.01, value=0),
         phase = FloatSliderWidget(min=-50, max=150, step=10, value=0), 
         data_flux=fixed(2), data_flux_err=fixed(0.1));
```
